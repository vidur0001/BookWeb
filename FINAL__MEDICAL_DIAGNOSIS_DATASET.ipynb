````python
# Write your code to make your dataset analysis ready.
#DataFrame remains unchanged because there are no missing values to fill.

data=df.fillna(method="bfill")
````

````python
# Convert 'Date of Admission' to datetime
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
````

````python
# Create a new column for patient age group
bins = [0, 18, 35, 50, 65, 100]
labels = ['0-17', '18-34', '35-49', '50-64', '65+']
df['age group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
````

````python
# Filter rows based on a condition (e.g., age greater than 50)
df_filtered = df[df['Age'] > 50]
````

````python
# Group by 'medication' and calculate the average billing amount
df_medication_billing = df.groupby('Medication')['Billing Amount'].mean().reset_index()
````

````python
# Pivot table to summarize data by 'gender' and 'admission type'
df_pivot = df.pivot_table(index='Gender', columns='Admission Type', values='Billing Amount', aggfunc='mean')
````

````python
# Convert categorical variables to numerical using Label Encoding
label_encoder = LabelEncoder()
df['gender_encoded'] = label_encoder.fit_transform(df['Gender'])
````

````python
# Normalize the 'billing amount' column
scaler = MinMaxScaler()
df['billing amount_normalized'] = scaler.fit_transform(df[['Billing Amount']])
````

````python
# Split the 'doctor' column into 'first name' and 'last name'
df[['doctor_first_name', 'doctor_last_name']] = df['Doctor'].str.split(' ', 1, expand=True)
````

````python
# Convert 'test results' to lowercase
df['Test Results'] = df['Test Results'].str.lower()
````

````python
# Rename columns for better readability
df.rename(columns={'Date of Admission': 'admission_date', 'Billing Amount': 'billing_amount'}, inplace=True)
````

````python
# Number of customers by insurance company

# Count the number of customers for each insurance company
insurance_counts = df['Insurance Provider'].value_counts()

# Get the insurance company with the most customers
most_customers = insurance_counts.idxmax()
most_customers_count = insurance_counts.max()

# Get the insurance company with the least customers
least_customers = insurance_counts.idxmin()
least_customers_count = insurance_counts.min()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=insurance_counts.index, y=insurance_counts.values, palette='viridis')
plt.title('Number of Customers by Insurance Company')
plt.xlabel('Insurance Company')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)

# Annotate the plot with the most and least customers
plt.text(insurance_counts.index.get_loc(most_customers), most_customers_count, f'Most: {most_customers_count}', ha='center', va='bottom')
plt.text(insurance_counts.index.get_loc(least_customers), least_customers_count, f'Least: {least_customers_count}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
````

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
# Handling Outliers & Outlier treatments
# Select only numerical columns for quartile calculation
numerical_df = df.select_dtypes(include=['int64', 'float64'])

# Calculate the first quartile (25th percentile)
Q1 = numerical_df.quantile(0.25)

# Calculate the third quartile (75th percentile)
Q3 = numerical_df.quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Calculate the lower bound for outliers
outlier1 = Q1 - 1.5 * IQR

# Calculate the upper bound for outliers
outlier2 = Q3 + 1.5 * IQR

# Count the number of outliers for each numerical column
num_outliers = ((numerical_df < outlier1) | (numerical_df > outlier2)).sum()

# Print the number of outliers for each column
print("Number of outliers for each numerical column:")
print(num_outliers)
````

We have used Inter Quartile Range(IQR) Technique, which gives us idea, which putliers are showing different from your Dataset.
